```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("MonthlyJJAS_weather.csv")
df
```

```python
# Extract the month from the 'date' column and convert it to a numeric value for sorting
df['month'] = pd.to_datetime(df['date']).dt.month
df
```

```python
# Create a new column for month names
month_names = {
    1: 'jan', 2: 'feb', 3: 'mar', 4: 'apr', 5: 'may', 6: 'jun', 7: 'jul', 8: 'aug', 9: 'sep', 10: 'oct', 11: 'nov', 12: 'dec'
}

```

```python
df['month_name'] = df['month'].map(month_names)
df
```

```python
# Pivot the DataFrame to transform each row data into column data for multiple value columns
pivot_df = df.pivot_table(index='year', columns='month_name', values=['prcp (mm/day)', 'srad (W/m^2)', 'swe (kg/m^2)', 'tmax (deg c)', 'tmin (deg c)', 'vp (Pa)'])

# Flatten the multi-level column index
pivot_df.columns = [f"{col[1]}_{col[0]}" for col in pivot_df.columns]

print(pivot_df)
```

```python
pivot_df.to_csv("Monthly_Weather_pivoted.csv")
```
