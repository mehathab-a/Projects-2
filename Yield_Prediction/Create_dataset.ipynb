```python
import pandas as pd
```

```python
dataset = pd.read_csv("Dataset/FAOSTAT_data_en_8-2-2023 (2).csv")
dataset
```

```python
cols = ['Element', 'Year', 'Value']
col_dataset = dataset[cols]
col_dataset
```

```python
area = col_dataset[col_dataset['Element']=='Area harvested']['Value']
area
```

```python
yields = col_dataset[col_dataset['Element']=='Yield']['Value']
yields
```

```python
new_df = pd.DataFrame({'Year':range(1980,2022), 'Yield':list(yields), 'Area Harvested':list(area)})
new_df
```

```python
new_df.to_csv('Yield_dataset.csv',index=False)
```
