```python
import numpy as np
import pandas as pd
```

```python
datasets = ['Dataset/Original_weather/Weather'+str(i)+'.csv' for i in [1,2,3,4,5]]
datasets
```

```python
# Initialize an empty list to store the DataFrames from each dataset
dataframes_list = []

for i,datas in enumerate(datasets):
    dfq = pd.read_csv(datas)
    print(dfq)
    
    df = dfq[(dfq['yday'] > 153) & (dfq['yday'] < 273)] 
    print(df)
    
    
    df['date'] = pd.to_datetime(df['year'].astype(str) + '-' + df['yday'].astype(str), format='%Y-%j')
    
    # Set 'date' as the index for easy resampling
    df.set_index('date', inplace=True)
    
    
    
    # Resample the data to monthly means
    df_monthly1 = df.drop(['prcp (mm/day)'],axis=1).resample('M').mean()
    df_monthly2 = df['prcp (mm/day)'].resample("M").sum()
    df_monthly = pd.concat([df_monthly1, df_monthly2], axis=1)
    
    
    # Drop the 'year' and 'yday' columns, as they are no longer needed
#     df_monthly = df_monthly.drop(columns=['year', 'yday'])
    
    # Droping all nan values
    df_monthly.dropna(inplace=True)
    
#     df_monthly.to_csv("New_Weather"+str(i)+".csv")
    # Append the current dataset's monthly means to the list
#     if 
    dataframes_list.append(df_monthly)

# Concatenate all the DataFrames from the list into a single DataFrame
aggregated_df = pd.concat(dataframes_list)

# Group the data by 'date' and calculate the mean across rows with similar dates
overall_means = aggregated_df.groupby('date').mean()

print(overall_means)

# Save the overall means to a CSV file
overall_means.to_csv("MonthlyJJAS24_weather.csv")
```
