```python
import os
import cv2
from skimage.transform import resize
from skimage.util import random_noise
import matplotlib.pyplot as plt
import numpy as np
```

```python
root_dir = 'dataset/'
dest_dir = 'modfied_dataset/'
```

```python
folder_names = os.listdir(root_dir)
print(folder_names)
```

```python
for i,folder in enumerate(folder_names):
    current_folder = root_dir+folder
    print(current_folder)
    dest_folder = dest_dir+folder+"/"
    image_names = os.listdir(current_folder)
    count = 1
    for j,img in enumerate(image_names):
        image_path = current_folder + "/"+img
        print(image_path)
        if img[-4:] == '.jpg':
            exten = '.jpg'
        elif img[-5:] == '.wepb':
            exten = 'wepb'
        else:
            exten = '.jpeg'
        
        
        
        
        # Read Image
        image_read = plt.imread(image_path)
        
        # Resize Images
        image_resized = resize(image_read,(299,299,3))
        dest_path = dest_folder+str(count)+exten
#         print(dest_path)
        plt.imsave(dest_path,image_resized)
        
        #Inject random noise in the image
        image_noised = random_noise(image_resized, mode='s&p', amount=0.0035)
        image_noised = np.array(255*image_noised, dtype='uint8')
        
        # Rotate Images by 90 degree
        image_90 = cv2.rotate(image_noised, cv2.ROTATE_90_CLOCKWISE)
        dest_path = dest_folder+str(count)+"_90"+exten
#         print(dest_path)
        plt.imsave(dest_path,image_90)
        
#         Rotate Images by 270 degree
#         image_270 = cv2.rotate(image_noised, cv2.ROTATE_90_COUNTERCLOCKWISE)
        dest_path = dest_folder+str(count)+"_noise"+exten
#         print(dest_path)
        plt.imsave(dest_path,image_noised)
        
#         
#         dest_path = dest_folder+str(count)+"_noise"+exten
#         print(dest_path)
#         plt.imsave(dest_path,image_noised) 
        
        #Gaussian Blur to image
        image_gaussian = cv2.GaussianBlur(image_resized, (5,5), 1)
        dest_path = dest_folder+str(count)+"_blur"+exten
#         print(dest_path)
        plt.imsave(dest_path,image_gaussian)
        
        # Gamma Correction
        image_cv2_read = cv2.imread(image_path)
        image_cv2 = cv2.resize(image_cv2_read, (299,299))
        invGamma = 1 / 2.25

        table = [((i / 255) ** invGamma) * 255 for i in range(256)]
        table = np.array(table, np.uint8)

        image_gamma = cv2.LUT(image_cv2, table)
        image_gamma = cv2.cvtColor(image_gamma, cv2.COLOR_BGR2RGB)
        dest_path = dest_folder+str(count)+"_gamma"+exten
#         print(dest_path)
        plt.imsave(dest_path,image_gamma)
        
        
        count += 1
        
```
